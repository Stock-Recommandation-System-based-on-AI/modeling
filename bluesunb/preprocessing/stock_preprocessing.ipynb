```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
df_origin = pd.read_csv('../data/samsung-stock-data-2000-2021.csv')
df_origin
```

```python
df_origin = df_origin.drop(['date', 'code', 'name', 'section'], axis=1)
```

```python
fig = plt.figure(figsize=(15,51))
n = len(df_origin.columns)
for i, c in enumerate(df_origin.columns):
    ax1 = fig.add_subplot(n,1,i+1)
    ax1.plot(df_origin[c], label=str(c))
    ax1.set_xticks(range(0, df_origin.shape[0], 732))
    ax1.set_ylabel(str(c), size=15)
    ax1.legend(loc='upper left', fontsize=12)
```

```python
df = df_origin.rolling(20).mean()
```

```python
fig = plt.figure(figsize=(15,51))
n = len(df.columns)
for i, c in enumerate(df.columns):
    ax1 = fig.add_subplot(n,1,i+1)
    ax1.plot(df[c], label=str(c))
    ax1.set_xticks(range(0, df.shape[0], 732))
    ax1.set_ylabel(str(c), size=15)
    ax1.legend(loc='upper left', fontsize=12)
```
